```python
# str 접근자 - replace(), split()
import pandas as pd

df = pd.DataFrame({
    'A': ['데이터 분석', '기본 학습서', '퇴근 후 열공'],
    'B': [10, 20, 30],
    'C': ['ab cd', 'AB CD', 'ab cd ']
})
df

# 문자열 변경 - str 사용
df['A'] = df['A'].str.replace("분석", "시각화")
# 숫자는 변경됨
df['B'] = df['B'].replace(20, 100)
df
```

```python
# str.split()
# df['A'].str.split(" ")
# df['A'] = df['A'].str.split(" ").str[1]

df['D'] = df['A'].str.split().str[0]
df
```

```python
# 특정 문자열 검색 - str.contains()
df["학습서포함유무"] = df['A'].str.contains("학습서")
df

# 문자의 길이 - str.len()
df["문자길이"] = df['A'].str.len()
df

# 소문자 - str.lower(). 대문자 - str.upper()
df['C'] = df['C'].str.lower()
df['C'] = df['C'].str.replace(" ", "")
df
```

```python
# 판다스 내장 함수
df = pd.read_csv("./data/cafe2.csv")
df

print("최대값: ", df['가격'].max())
print("최소값: ", df['가격'].min())
print("평균값: ", df['가격'].mean())
print("중앙값: ", df['가격'].median())
print("합계: ", df['가격'].sum())
print("표준편차: ", df['가격'].std())  #표준편차
print("분산: ", df['가격'].var())  #분산
print(df)
print("최대값위치: ", df['가격'].idxmax())

# 사분위수 - quantile(.25)
print("1사분위수: ", df['가격'].quantile(.25))
print("3사분위수: ", df['가격'].quantile(.75))

# 1사분위수 보다 작은 가격 데이터 출력
df[df['가격'].quantile(.25) > df['가격']]
```

```python
import pandas as pd
# 그룹핑 - groupby()
df = pd.read_csv("./data/cafe2.csv")
df

df['원산지'] = ["베트남", "과테말라", "베네수엘라", "베트남", "베트남", "과테말라", "과테말라"]
# df

# df.loc[6, "칼로리"] = 210.0
df.iloc[6, 2] = 210.0
df

# "원산지" 기준으로 그룹화
# df.groupby(['원산지']).mean(numeric_only=True)

# 원산지와 칼로리로 그룹화
df.groupby(['원산지', "칼로리"]).mean(numeric_only=True)

# agg() - 집계 연산
df.groupby(['원산지', "메뉴"]).agg(["mean", "sum"])

# cafe3으로 저장
df.to_csv("./data/cafe3.csv", index=False)

df = pd.read_csv("./data/cafe3.csv")
df
```

```python
# 시간, 날짜 자료형 - 시계열 데이터
# 문자열(24-11-15) - object -> datetime
# to_datetime()

import pandas as pd

data = {
    'Date1': ["2024-11-10", "2024-11-11", "2024-11-12", "2024-11-13"],
    'Date2': ["2024/11/10", "2024/11/11", "2024/11/12", "2024/11/13"],
    'DateTime1': ["2024-11-10 10:10:10", "2024-11-11 10:20:20", 
                  "2024-11-12 11:30:30", "2024-11-13 11:40:40"],
    'DateTime2': ["2024-11-10 10-10-10", "2024-11-11 10-20-20", 
                  "2024-11-12 11-30-30", "2024-11-13 11-40-40"]
}

df = pd.DataFrame(data)
df

df.info()

df['Date1'] = pd.to_datetime(df['Date1'])
df['Date2'] = pd.to_datetime(df['Date2'])
df['DateTime1'] = pd.to_datetime(df['DateTime1'])
df['DateTime2'] = pd.to_datetime(df['DateTime2'], format="%Y-%m-%d %H-%M-%S")
df.info()

# year, month, day 파생 칼럼 . dt 속성
df['year'] = df['DateTime1'].dt.year
df['month'] = df['DateTime1'].dt.month
df['day'] = df['DateTime1'].dt.day
df['hour'] = df['DateTime1'].dt.hour
df['minute'] = df['DateTime1'].dt.minute
df['second'] = df['DateTime1'].dt.second

# 요일(0-월, 1-화, 2-수... 5-토,6-일)
day_of_week = df['DateTime1'].dt.dayofweek
print(day_of_week)
df

# date_data.csv 저장
df.to_csv("./data/date_data.csv", index=False)
```

```python
import pandas as pd
# Timedelta - 시간의 차이를 구현
df = pd.read_csv("./data/date_data.csv", usecols=['DateTime1'])
df
df.info()

df['DateTime1'] = pd.to_datetime(df['DateTime1'])

# 10일후
day = pd.Timedelta(days=9)
df['10day'] = df['DateTime1'] + day

# 10시간 후
hour = pd.Timedelta(hours=9)
df['10hour'] = df['DateTime1'] + hour

df.info()
df
```

```python
# 연습문제
import pandas as pd

df = pd.read_csv("./data/type1_data1.csv")

# 자료형 변환
df["subscribed"] = pd.to_datetime(df["subscribed"])

# 월별 칼럼 생성
df['month'] = df["subscribed"].dt.month

# 월별 데이터 합계
df = df.groupby('month').sum(numeric_only=True)

# 'views' 오름차순 정렬
df = df.sort_values("views")
print(df.index[0])

df.info()
df.head()
```

```python
# 21
# dt.total_seconds()
import pandas as pd

df = pd.read_csv("./data/delivery_time.csv")
df

# 실제도착시간, 예상도착시간 자료형 변환
df['실제도착시간'] = pd.to_datetime(df['실제도착시간'])
df['예상도착시간'] = pd.to_datetime(df['예상도착시간'])

# 지연시간 계산 - 분단위로 환산 : 초의 총합 / 60
df['지연시간'] = (df['실제도착시간'] - df['예상도착시간']).dt.total_seconds() / 60

# 조건1 - 예상도착시간보다 늦게 도착
cond1 = df['지연시간'] > 0
print(len(df[cond1]))

# 조건2 - 거리가 7km 이상인 데이터 수
cond2 = df['거리'] > 7
print(len(df[cond2]))

# print(df[cond1 & cond2])

# 데이터 수
print(len(df[cond1 & cond2]))

# df.info()
# df.head()
```

```python
# 데이터 프레임 합치기
import pandas as pd

# 에피타이저 메뉴
appetizer = pd.DataFrame({
    'Menu': ['Salad', 'Soup', 'Bread'],
    'Price': [5000, 3000, 2000]
})

# 메인 메뉴
main = pd.DataFrame({
    'Menu': ['Steak', 'Pasta', 'Chicken'],
    'Price': [15000, 12000, 10000]
})
print(appetizer)
print(main)

# concat(프레임1, 프레임2)
full_menu = pd.concat([appetizer, main], axis=1)
full_menu
```

```python
# 메뉴와 가격
# 키값 기준으로 병합 - merge
price = pd.DataFrame({
    'Menu': ['Salad', 'Soup', 'Steak', 'Pasta'],
    'Price': [5000, 3000, 15000, 12000]
})

# 메뉴와 칼로리
cal = pd.DataFrame({
    'Menu': ['Soup', 'Steak', 'Pasta','Salad'],
    'Calories': [100, 500, 400, 150]
})

menu_info = pd.merge(price, cal, on="Menu")
menu_info
```
