```python
# 작업형 유형1
#01
import pandas as pd

df = pd.read_csv("./data/type1_data1.csv")
df.info()
df.head()

# df['f5'].value_counts()

# 0 제외
cond1 = (df['f5'] != 0)
print(df[cond1])

# not 연산자 
print(~df[cond1])

# 'views' 칼럼의 결측치 최소값으로 채우기
min_val = df['views'].min()
print(min_val)

df['views'] = df['views'].fillna(min_val)
df.info()

# 'views' 칼럼의 중앙값
print(int(df['views'].median()))
```

```python
# 3
import pandas as pd

df = pd.read_csv("./data/type1_data1.csv")

df.info()

# NaN의 개수 파악
df.isnull().sum()

df = df.dropna()

df.isnull().sum()

# 새 칼럼 추가
df['new'] = df['views'] / df['f1']

# 내림차순 정렬
df = df.sort_values('new', ascending=False)

# age 값
print(int(df.iloc[0, 1]))
df.head()
```
