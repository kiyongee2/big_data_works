### 은행 정보로 신용 등급 예측

```python
# 정답 칼럼 - Credit_Score
# f1-macro 평가지표 

import pandas as pd
# 데이터 준비 
train = pd.read_csv("score_train.csv")
test = pd.read_csv("score_test.csv")

# 데이터 탐색
# train.info()
train["Credit_Score"].value_counts()

# 요약 통계 
train.describe() #수치형 
train.describe(include="object") #수치형 

# 데이터 선택
target = train.pop("Credit_Score")
print(train.shape, test.shape)

# 인코딩(원-핫)
# train = pd.get_dummies(train)
# test = pd.get_dummies(test)
# print(train.shape, test.shape)

# 레이블 인코딩
cols = train.select_dtypes(include='O').columns
print(cols)
# cols = ["Payment_of_Min_Amount", "Credit_Mix", 'Payment_Behaviour']
# print(train.info())

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
le = LabelEncoder()
for col in cols:
    train[col] = le.fit_transform(train[col])
    test[col] = le.fit_transform(test[col])
# print(train.columns)

# 스케일링 - 수치형 칼럼 대상
cols2 = train.select_dtypes(include='number').columns
print(cols2)

scaler = MinMaxScaler()
train[cols2] = scaler.fit_transform(train[cols2])
test[cols2] = scaler.fit_transform(test[cols2])

print(train.head())
# 검증 데이터 분할
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train, 
                                                    target, 
                                                    test_size=0.2, 
                                                    random_state=42)
# help(train_test_split)

# 머신러닝 모델 학습 및 평가
# 의사결정 나무(Decision Tree) - max_detph, n_estimate
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
# 모델 생성
model = RandomForestClassifier(random_state=42)
# model = RandomForestClassifier(max_depth=3, n_estimators=300, random_state=42)
# 학습(훈련용)
model.fit(X_train, y_train)

# 예측값(검증용 데이터)
pred = model.predict(X_test)

# 평가(측정)
f1 = f1_score(y_test, pred, average='macro')
print("f1_score:", f1)

# test 데이터 평가 및 파일 제출
pred2 = model.predict(test)
pred2

submit = pd.DataFrame({"pred": pred2})
submit.to_csv("result.csv", index=False)

pd.read_csv("result.csv")
```
