```python
import pandas as pd

train = pd.read_csv("data/customer_train.csv")
test = pd.read_csv("data/customer_test.csv")

# print(train)
# print(test)

# print(train.info())

# 결측치 대체
# print(train.isnull().sum())
median = train['환불금액'].median()
# print(median)
train['환불금액'] = train['환불금액'].fillna(median)
# print(train.isnull().sum())

# 인코딩 - '주구매상품', '주구매지점'
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
train['주구매상품'] = encoder.fit_transform(train['주구매상품'])
train['주구매지점'] = encoder.fit_transform(train['주구매지점'])

test['주구매상품'] = encoder.fit_transform(test['주구매상품'])
test['주구매지점'] = encoder.fit_transform(test['주구매지점'])

# print(train.info())

# 데이터 분할(훈련/검증)
from sklearn.model_selection import train_test_split

# 데이터 선택
# target = train.pop('성별')
X = train.drop(columns=['성별'])
y = train['성별']

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)

# print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

# 모델 생성 및 학습, 평가
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=11)
model.fit(X_train, y_train)
pred1 = model.predict(X_test)
# print(pred1)

# 평가지표에 의한 성능 측정
from sklearn.metrics import roc_auc_score

score = roc_auc_score(y_test, pred1)
print("score:", score)

# 평가용 데이터 성능 측정
pred2 = model.predict(test)
print(pred2)

# 파일 제출
df = pd.DataFrame({"pred": pred2})
df.to_csv("result.csv", index=False)

print(pd.read_csv("result.csv"))
```
