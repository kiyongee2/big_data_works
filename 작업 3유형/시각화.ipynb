```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 30]

# 그래프 생성 
plt.plot(x, y)
plt.show()
```

```python
# 범례

a = [10, 20, 30, 40]
b = [40, 30, 20, 10]

# 한글글꼴
plt.rc('font', family="Malgun Gothic")

plt.title("기본 그래프")
plt.plot(a, label="demand")
plt.plot(b, label="supply")
plt.legend()
plt.xlabel("수량")
plt.ylabel("매출액")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 30]

# 그래프 생성 (bar-세로막대, barh-가로막대)
plt.barh(x, y)
plt.show()
```
